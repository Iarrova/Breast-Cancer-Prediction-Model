```python
import os

import pandas as pd
import matplotlib.pyplot as plt
```

# Load Datasets

```python
DATASET_DIR = "../datasets/CCBIS-DDSM/"
```

```python
train_df = pd.read_csv(os.path.join(DATASET_DIR, "train.csv"))
validation_df = pd.read_csv(os.path.join(DATASET_DIR, "validation.csv"))
test_df = pd.read_csv(os.path.join(DATASET_DIR, "test.csv"))
```

# Explore Bounding Box Sizes

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 10))
train_df["w"].hist(grid=False, ax=ax[0], bins=50)
train_df["h"].hist(grid=False, ax=ax[1], bins=50)
plt.tight_layout()
plt.show()

print("Mean Width: {} and mean height: {}".format(train_df["w"].mean(), train_df["h"].mean()))
print("Std Width: {} and std height: {}".format(train_df["w"].std(), train_df["h"].std()))
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 10))
validation_df["w"].hist(grid=False, ax=ax[0], bins=50)
validation_df["h"].hist(grid=False, ax=ax[1], bins=50)
plt.tight_layout()
plt.show()

print("Mean Width: {} and mean height: {}".format(validation_df["w"].mean(), validation_df["h"].mean()))
print("Std Width: {} and std height: {}".format(validation_df["w"].std(), validation_df["h"].std()))
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 10))
test_df["w"].hist(grid=False, ax=ax[0], bins=50)
test_df["h"].hist(grid=False, ax=ax[1], bins=50)
plt.tight_layout()
plt.show()

print("Mean Width: {} and mean height: {}".format(test_df["w"].mean(), test_df["h"].mean()))
print("Std Width: {} and std height: {}".format(test_df["w"].std(), test_df["h"].std()))
```
